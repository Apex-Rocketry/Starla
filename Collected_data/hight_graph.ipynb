```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("bme_data.csv")
```

```python
altitude = data["altitude"]
z_velocity = data["z_velocity"]
time = data["time"]
```

```python
plt.axis([0, 20, 20, 22.5])
plt.plot(time, altitude, time, z_velocity)
plt.show()
```
