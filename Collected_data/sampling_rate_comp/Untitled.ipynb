```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
linear_data = pd.read_csv("data.csv")
threaded_data = pd.read_csv("beta_data.csv")
```

```python
linear_sr = linear_data["sr_list"]
linear_time = linear_data["time_list"]

threaded_sr = threaded_data["sr_list"]
threaded_time = threaded_data["time_list"]
```

```python
plt.axis([2, 8,0, 0.06])
plt.plot(threaded_time, threaded_sr)
plt.title('Threaded sr')
plt.show()
```

```python
plt.axis([3.5, 8,0, 0.06])
plt.plot(linear_time, linear_sr)
plt.title('Linear sr')
plt.show()
```
